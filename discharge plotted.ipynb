```python
import matplotlib.pyplot as plt
import math
print("enter  the dischargs of 2,5,10,20,and 50 years events")
items=[]
for I in range(0,5):
   value=float(input(" "))
   items.append(value)
a=sum(items)
av=a/5
years =[2,5,10,20,50]
def tcalc(t):
    return -math.log(-math.log(1-(1/t)))
sa=0
for p in range(0,5):
    sa=sa+tcalc(years[p])
avg = sa/5
da=0
ba=0
for m in range(0,5):
    da=da+(tcalc(years[m])-avg)*(items[m]-av)
    ba=ba+(tcalc(years[m])-avg)**2
alpha=da/ba
u=av-alpha*avg
print("\n Extrapolated Discharge for 200,500 and 1000 years floods are: ",round(u+alpha*tcalc(200),3) ,round(u+alpha*tcalc(500),3), round(u+alpha*tcalc(1000),3))
E=[items[0], items[1],items[2],items[3],items[4],round(u+alpha*tcalc(200),3) ,round(u+alpha*tcalc(500),3), round(u+alpha*tcalc(1000),3)]
P=[2,5,10,20,50,200,500,1000]
plt.plot(P,E)
plt.title("Extraploatting discharges")
```
